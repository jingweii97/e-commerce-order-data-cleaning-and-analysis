```python
import pandas as pd
df = pd.read_excel('DummyData\IncomeData\Shopee\shopee_dummy_income_released.xlsx',sheet_name='Income', header = 5)
df
```

```python
df.drop(columns = 'Sequence No.',inplace=True)
df
```

```python
df.info()
```

```python
# Cast date time accordingly
date_columns = df.columns[df.columns.str.lower().str.contains('date')]
print(list(date_columns))
df[date_columns] = df[date_columns].apply(pd.to_datetime)
df.info()
```

```python
df
```

```python
from sqlalchemy import create_engine
server_name = 'MSI\SQLEXPRESS'
driver = 'ODBC Driver 17 for SQL Server'
database = 'Kaizen_db'
table_name = 'Shopee_Income'
engine = create_engine(f"mssql+pyodbc://@{server_name}/{database}?driver={driver}&trusted_connection=yes")
try:
    df.to_sql(table_name, engine, if_exists='append', index=False)
    print('Data appended to server successfully')
except Exception as e:
    print(f'Error occured {e}')
```
