```python
import pandas as pd
df = pd.read_csv('DummyData\IncomeData\Lazada\lazada_dummy_income_released.csv', sep = ';')
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
df
```

```python
df.info()
```

```python
# Drop column of Fee Name
df2 = df.drop(columns = ['Fee Name','Comment'])
df2
```

```python
# Group the data by orders
grouping_columns = list(df2.select_dtypes(exclude='number').columns) + ['Order Line ID']
group_by_order = df2.groupby(list(grouping_columns))
group_by_order = group_by_order.sum()
group_by_order = group_by_order.reset_index()
group_by_order
```

```python
from sqlalchemy import create_engine
server_name = 'MSI\SQLEXPRESS'
driver = 'ODBC Driver 17 for SQL Server'
database = 'Kaizen_db'
table_name = 'Lazada_Income'
engine = create_engine(f"mssql+pyodbc://@{server_name}/{database}?driver={driver}&trusted_connection=yes")
try:
    group_by_order.to_sql(table_name, engine, if_exists='append', index=False)
    print('Data appended to server successfully')
except Exception as e:
    print(f'Error occured {e}')
```
