# Cleanup imported outbound data from fulfillment

```python
import pandas as pd
```

```python
# Import the order data
filepath =  'DummyData\OrderData'
filename = 'dummy_order_data.xlsx'
df = pd.read_excel(filepath+'\\'+filename, dtype={'Marketplace Order ID': str})
df
# print(df['Marketplace Order ID'].dtype) 
```

```python
df.info()
```

```python
# Creating new columns to cater for exploded values
df["SKU"] = ""
df["Quantity"] = 0
df
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
df = df.drop(['No'], axis = 1)
df
```

```python
# Cast it to string
df['Product Quantity'] = df['Product Quantity'].astype(str)
```

```python
# Explode the skus and respective quantities, into rows
row_list = []
for index,row in df.iterrows():
    skus = row['Product Seller SKU'].split(',')
    quantity = row['Product Quantity'].split(',')
    
    for sku,quantity in zip(skus,quantity):
        new_row = row.copy()
        new_row['SKU'] = sku.strip()
        new_row['Quantity'] = quantity.strip()
        row_list.append(new_row)

transformed_df = pd.DataFrame(row_list)
transformed_df
```

```python
# Reset the index after transforming the data
transformed_df.reset_index(drop=True, inplace = True)
```
